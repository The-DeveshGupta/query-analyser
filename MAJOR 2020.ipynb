# Image Processing

### Import Library

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

image = cv2.imread('sample2.jpg',0)
image
plt.imshow(image,'gray')
```

### Resize Image

```python
print('Original Dimensions : ',image.shape)

mul = 4
width = int(image.shape[1] * mul)
height = int(image.shape[0] * mul)
dim = (width, height)

resized = cv2.resize(image, dim, interpolation = cv2.INTER_AREA)
 
print('Resized Dimensions : ',resized.shape)
```

```python
image = resized
```

# Natural Language Processing

```python
print(txt)
```

```python
text = ""
for item in txt:
    text = text + " " + item
text
```

```python
words = text.split()
text = " ".join(sorted(set(words), key=words.index))
text
```

```python
from nltk.tokenize import sent_tokenize

tokenized_text=sent_tokenize(text)
print(tokenized_text)
```

```python
from nltk.tokenize import word_tokenize
tokenized_word=word_tokenize(text)
print(tokenized_word)
```

```python
from nltk.probability import FreqDist
fdist = FreqDist(tokenized_word)
print(fdist)
```

```python
fdist.most_common(2)
```

```python
# Frequency Distribution Plot
import matplotlib.pyplot as plt
fdist.plot(30,cumulative=False)
plt.show()
```

```python
from nltk.corpus import stopwords
stop_words=set(stopwords.words("english"))
print(stop_words)
```

```python
filtered_sent=[]
for w in tokenized_word:
    if w not in stop_words:
        filtered_sent.append(w)
print("Tokenized Sentence:",tokenized_word)
print("Filterd Sentence:",filtered_sent)
```

```python
from nltk.stem import PorterStemmer
from nltk.tokenize import sent_tokenize, word_tokenize

ps = PorterStemmer()

stemmed_words=[]
for w in filtered_sent:
    stemmed_words.append(ps.stem(w))

print("Filtered Sentence:",filtered_sent)
print("Stemmed Sentence:",stemmed_words)
```

```python
import nltk
from nltk.stem.wordnet import WordNetLemmatizer
nltk.download('wordnet')
lem = WordNetLemmatizer()

from nltk.stem.porter import PorterStemmer
stem = PorterStemmer()

lemmatize_sent = ""
stemmed_sent = ""
for w in stemmed_words:
    lemmatize_sent = lemmatize_sent + " " + lem.lemmatize(w,"v")
    stemmed_sent = stemmed_sent + " " + stem.stem(w)

print("Lemmatized Word:",lemmatize_sent)
print("Stemmed Word:",stemmed_sent)
```

```python
query = lemmatize_sent
query
```
